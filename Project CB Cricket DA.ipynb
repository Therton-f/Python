```python
import pandas as pd
import json
```

```python
with open(r"C:\Users\Therton_F\Downloads\DA\Python\Project C Cricket\t20_json_files\t20_wc_match_results.json") as f:
    data = json.load(f)
data
```

```python
#all data is stored in one array, within that match summary is the one we wanted
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.rename({'scorecard':'match_id'}, axis=1, inplace=True)
df_match.head()
```

```python
match_ids_dict = {} #we have created dictionaries for each match by giving them match id and VS title

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']

match_ids_dict
```

```python
with open(r"C:\Users\Therton_F\Downloads\DA\Python\Project C Cricket\t20_json_files\t20_wc_batting_summary.json") as f:
    data = json.load(f)

    all_records = []
    for rec in data:
        all_records.extend(rec['battingSummary'])

all_records

df_batting = pd.DataFrame(all_records)
df_batting.tail()
```

```python
df_batting['out/not_out'] = df_batting.dismissal.apply(lambda x: 'out' if len(x) > 0 else 'not_out')
df_batting.head()
```

```python
df_batting.drop(columns=['dismissal'], inplace=True)
df_batting
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting.head()
```
