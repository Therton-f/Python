### Scrapping Table from Website

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'

page = requests.get(url)

soup2 = BeautifulSoup(page.text, 'html')
```

```python
soup2.find_all('table')
```

```python
#there more than 1 table here, from looking at output we can see, the table we want is at 2nd i.e. index=1
table = soup2.find_all('table')[1]
print(table)
```

```python
soup2.find_all('table', class_ = 'wikitable sortable')

#query above and below gives the same output, those are 2 ways to do it
#<table class="wikitable sortable jquery-tablesorter">
```

```python
world_titles = table.find_all('th')
world_titles
```

```python
#extracting title values and storing in a list with use of for loop

world_table_titles = [title.text.strip() for title in world_titles]

print(world_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_table_titles)
df
```

```python
#we have already html code for table 1, from that using'th' we got titles, now from 'tr' rows & 'td' datas within rows
```

```python
table.find_all('tr')
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data:
    print(row.find_all('td'))
```

```python
for row in column_data:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print(individual_row_data)
```

```python
#starting from 1 bcz 1st list is null as seen on above output

for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data
```
