```python
days =  [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title("Weather")

plt.plot(days,max_t, label='Max')
plt.plot(days,min_t, label='Min')
plt.plot(days,avg_t, label='Avg')

plt.legend(loc="best", shadow=True, fontsize="small")  #giving individual line chart a label name then calling through legend
                                                       #legend(loc="lower right") or you can write "best"
plt.grid()
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
plt.bar(company,revenue)
```

```python
#sometimes bar chart wont take string values in x axis
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue)
```

```python
#now with xticks we will give array values from corresponding company name
#w/ this 0 will be displayed as Googl, same for 1,2,3
plt.xticks(xpos,company)

plt.ylabel("revenue(bin)")
plt.title("US Tech Stocks")
plt.bar(xpos,revenue, label="Revenue")
plt.bar(xpos,profit, label="Profit")

plt.legend()
```

```python
plt.xticks(xpos,company)

plt.ylabel("revenue(bin)")
plt.title("US Tech Stocks")
plt.bar(xpos-0.2,revenue, width=0.4,label="Revenue")
plt.bar(xpos+0.2,profit, width=0.4,label="Profit")

plt.legend()
```

```python
plt.yticks(xpos,company)

plt.title("US Tech Stocks")
plt.barh(xpos-0.2,revenue, label="Revenue")
plt.barh(xpos+0.2,profit, label="Profit")

plt.legend()
```

```python
blood_sugar_men = [113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women = [67,98,89,120,133,150,84,69,89,79,120,112,100]
```

```python
plt.hist(blood_sugar_men, bins=4, rwidth=0.95)
#by default 10 bins
#rwidth is width of bar and by default all bars are conected with no space
```

```python
#instead of bin size we can give actual numbers
plt.hist(blood_sugar_men, bins=[80,100,125,150], rwidth=0.95, color='g', histtype='step')
```

```python
plt.xlabel('sugar range')
plt.ylabel('Total no of patients')
plt.title('Blood sugar analysis')
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'], label=['men','women'])
plt.legend()
```

```python
plt.xlabel('sugar range')
plt.ylabel('Total no of patients')
plt.title('Blood sugar analysis')
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'], label=['men','women'],orientation='horizontal')
plt.legend()
#but with horiontal orientation we have change x & y label
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car","Other Utilities"]
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, radius=1.5, autopct='%0.2f%%', shadow=True)
 #by default radius is 1 #0.0f means 0 decimal place, 0.1 means 1 decimal place
plt.show() #to remove values from output
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, radius=1.5, autopct='%0.2f%%', explode=[0,0.2,0,0,0])
 #explode carving out pie by given distance
plt.show()
```

```python
#by default pie1st pie start from 0 degree and goes counterclockwise
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, radius=1.5, autopct='%0.2f%%', startangle=45)
plt.show()

plt.savefig('piechart.jpeg', bbox_inches="tight", pad_inches=2, transparent=False)
#bbox used to fit the chart in frame
#pad_inches provides extra padding outside figure
#transparent=true means background is fully transparent
```
